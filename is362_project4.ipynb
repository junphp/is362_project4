```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
#from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.cross_validation import train_test_split
```

** Start with the mushroom data in the pandas DataFrame that you constructed in your Assignment – Preprocessing Data with sci-kit learn. **

```python
Location = 'https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data'
data = pd.read_csv(Location, usecols=[0,2,5],names=["edibility","cap-surface","odor"])
data.head()
```

```python
data["edibility"].replace(['e', 'p'], [0, 1], inplace=True)
data["cap-surface"].replace(['f','g','y','s'],[0,1,2,3], inplace=True)
data["odor"].replace(['a', 'l', 'c', 'y', 'f', 'm', 'n', 'p', 's'], [0,1,2,3,4,5,6,7,8], inplace=True)
```

```python
data.head(10)
```

**Use scikit-learn to determine which of the two predictor columns that you selected (odor and one other column of your choice) most accurately predicts whether or not a mushroom is poisonous.  There is an additional challenge here—to use scikit-learn’s predictive classifiers, you’ll want to convert each of your two (numeric categorical) predictor columns into a set of columns.  See for one approach pandas get_dummies() method. **

```python
cap_data = pd.Series(data["cap-surface"])
d_cap_data = pd.get_dummies(cap_data)

od_data = pd.Series(data["odor"])
d_od_data = pd.get_dummies(od_data)

e_data = pd.Series(data["edibility"])
d_e_data = pd.get_dummies(e_data)

frames = [d_cap_data, d_od_data, d_e_data]
```

```python
pd.concat(frames,axis = 1)
```

```python
data_frames = pd.concat(frames,axis = 1)
```
